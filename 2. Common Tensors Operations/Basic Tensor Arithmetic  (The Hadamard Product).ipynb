# No 1

```python
import numpy as np

# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[12, 24, 36], [16, 32, 48]])
print(X)
```

```python
import torch as pt

# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = pt.tensor([[13, 26, 39], [15, 30, 45]])
print(X_pt)
```

```python
import tensorflow as tf

# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[14, 28, 42], [18, 36, 54]])
print(X_tf)
```

```python
# lakukan perkalian matrix X, X_pt dan X_tf dengan suatu skalar
X*2, X_pt*2, X_tf*2
```

```python
# lakukan penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar
X+8, X_pt+8, X_tf+8
```

```python
# lakukan pengurangan matrix X, X_pt dan X_tf dengan suatu skalar
X-3, X_pt-3, X_tf-3
```

# No 2

```python
# buat matrix Y dengan ukuran 2x3 menggunakan numpy
Y = np.array([[23, 32, 45], [42, 54, 67]])
print(Y)
```

```python
# buat matrix Y_pt dengan ukuran 2x3 menggunakan pytorch
Y_pt = pt.tensor([[24, 33, 46], [43, 55, 68]])
print(Y_pt)
```

```python
# buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow
Y_tf = tf.Variable([[25, 34, 47], [44, 56, 69]])
print(Y_tf)
```

```python
# kalikan matrix X dengan Y
X*Y
```

```python
# kalikan matrix X_pt dengan Y_pt
X_pt*Y_pt
```

```python
# kalikan matrix X_tf dengan Y_tf
X_tf*Y_tf
```
