```python
import numpy as np

# buat vector x dan y dengan ukuran 1x3 menggunakan numpy
x = np.array([14, 28, 63])
y = np.array([43, 95, 67])
print(f"Vector x : {x}")
print(f"Vector y : {y}")
```

```python
import torch as pt

# buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
x_pt = pt.tensor([44, 68, 23])
y_pt = pt.tensor([48, 65, 57])
print(f"Vector x_pt : {x_pt}")
print(f"Vector y_pt : {y_pt}")
```

```python
import tensorflow as tf

# buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
x_tf = tf.Variable([46, 48, 63])
y_tf = tf.Variable([68, 35, 52])
print(f"Vector x_tf : {x_tf}")
print(f"Vector y_tf : {y_tf}")
```

# No 2

```python
# hitung perkalian dot product antara vector x dan vector y
np.dot(x, y)
```

```python
# hitung perkalian dot product antara vector x_pt dan vector y_pt
pt.dot(x_pt, y_pt)
```

```python
# hitung perkalian dot product antara vector x_tf dan vector y_tf
tf.tensordot(x_tf, y_tf, axes=1)
```

```python
# Nomor 1 - Apa Transpose dari matrix Y?
Y = np.array([[42, 4, 7, 99], [-99, -3, 17, 22]])
print(f"Matrix Y : \n{Y}")
Y_t = Y.T
print(f"Matrix Y Transpose : \n{Y_t}")
```

```python
# Nomor 2 - Apa Hasil Hadamard Product antara matrix A dan B

A = np.array([[25, 10], [-2, 1]])
B = np.array([[-1, 7], [10, 8]])
print(f"Matrix A : \n{A}")
print(f"Matrix B : \n{B}")

hadamard_product = A * B
print(f"Hadamard Product antara A dan B : \n{hadamard_product}")
```
