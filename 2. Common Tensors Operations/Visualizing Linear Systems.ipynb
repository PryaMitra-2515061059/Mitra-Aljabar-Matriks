```python
# buat visualisasi dari persamaan berikut : y = 2x+1 dan 3x-y = 5

import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

y1 = 2 * x + 1

y2 = 3 * x - 5

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = 2x + 1')
plt.plot(x, y2, label='3x - y = 5')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Visual dari Persamaan Linear')
plt.legend()
plt.grid(True)
plt.show()
```
