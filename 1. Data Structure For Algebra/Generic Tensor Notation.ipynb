```python
import torch as pt
import tensorflow as tf

# jelaskan tentang perintah ini

images_pt = pt.zeros([32, 28, 28, 3])
images_pt

# Penjelasan

# 1. pt.zeros
# Fungsi ini digunakan untuk membuat tensor yang berisi nilai nol. Dalam hal ini, kita membuat tensor dengan ukuran [32, 28, 28, 3], yang berarti kita memiliki 32 gambar dengan ukuran 28x28 piksel dan 3 saluran warna (RGB).

# 2. [32, 28, 28, 3]
# Ini adalah bentuk array yang menentukan dimensi tensor. Dalam hal ini:
# - 32: jumlah gambar dalam batch
# - 28: tinggi dan lebar setiap gambar (dalam piksel)
# - 3: jumlah saluran warna (misalnya RGB)

# Jadi, perintah ini membuat sebuah tensor yang berisi nilai nol dengan ukuran yang sesuai untuk menyimpan 32 gambar berukuran 28x28 piksel dengan 3 saluran warna.
# itulah kenapa kita menggunakan pt.zeros untuk membuat tensor yang berisi nilai nol, karena kita ingin menginisialisasi tensor dengan nilai awal sebelum mengisi dengan data gambar yang sebenarnya.
```
