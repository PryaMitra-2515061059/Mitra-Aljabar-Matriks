```python
import numpy as np
import torch as pt
import tensorflow as tf
```

# No 1

```python
# buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy

A = np.array([[14, 62, 93], [42, 53, 60]])
b = np.array([[57], [58], [95]])
print(A)
print(b)
```

```python
# buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch

A_pt = pt.tensor([[33, 32, 95], [22, 63, 81]])
b_pt = pt.tensor([[67], [88], [95]])
print(A_pt)
print(b_pt)
```

```python
# buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
A_tf = tf.Variable([[83, 52, 45], [22, 66, 71]])
b_tf = tf.Variable([[70], [83], [90]])
print(A_tf)
print(b_tf)
```

# No 2

```python
# tampilkan ukuran matrix A dan vector b
print("Ukuran matrix A:", A.shape)
print("Ukuran vector b:", b.shape)
```

# No 3

```python
# kalikan matrix A dengan vector b
np.dot(A, b)
print("Hasil perkalian matrix A dan vector b : \n", np.dot(A, b))
```

```python
# kalikan matrix A_pt dengan vector b_pt
pt.matmul(A_pt, b_pt)
print("Hasil perkalian matrix A_pt dan vector b_pt : \n", pt.matmul(A_pt, b_pt))
```

```python
# kalikan matrix A_tf dengan vector b_tf
tf.linalg.matmul(A_tf, b_tf)
```
