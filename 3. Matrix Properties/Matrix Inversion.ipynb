```python
# selesaikan persamaan berikut menggunakan numpy, tensorflow dan pytorch
# 4b + 2c = 4
# -5b + -3c = -7
```

## Numpy

```python
# selesaikan persamaan berikut menggunakan numpy
# 4b + 2c = 4
# -5b + -3c = -7
A = np.array([[4, 2], [-5, -3]])
B = np.array([[4], [-7]])
C = np.linalg.solve(A, B)
print(C)
```

## Torch

```python
# selesaikan persamaan berikut menggunakan pytorch
# 4b + 2c = 4
# -5b + -3c = -7
A = pt.tensor([[4, 2], [-5, -3]], dtype=pt.float32)
B = pt.tensor([[4], [-7]], dtype=pt.float32)
C = pt.linalg.solve(A, B)
print(C)
```

## Tensorflow

```python
# selesaikan persamaan berikut menggunakan tensorflow
# 4b + 2c = 4
# -5b + -3c = -7
A = tf.Variable([[4, 2], [-5, -3]], dtype=tf.float64)
B = tf.Variable([[4], [-7]], dtype=tf.float64)
C = tf.linalg.solve(A, B)
print(C)
```
