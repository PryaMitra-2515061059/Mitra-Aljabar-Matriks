```python
import numpy as np
import torch as pt
import tensorflow as tf
```

### Numpy

```python
X = np.array([[17, 43, 85], [64, 24, 51], [11, 75, 77]])
print(X)
np.linalg.norm(X)
```

### Torch

```python
X_pt = pt.tensor([[13, 23, 55], [14, 24, 56], [15, 25, 57]])
X_pt = X_pt.float()
print(X_pt)
pt.norm(X_pt)
```

### Tensorflow

```python
X_tf = tf.Variable([[23, 63, 87], [15, 24, 33], [88, 75, 90]], dtype=tf.float32)
print(X_tf)
tf.norm(X_tf)
```
